```python
# function to define trend based on Day_perc_change where n is the parameter
def trend(n):
    if(n < 0.5 and n >-0.5):
        print('\'Slight or no change\'' + ',' )
    elif(n < 1 and n >= .5):
        print("\'Slight positive\'" + ',' )
    elif(n <= -.5 and n > -1):
        print('\'Slight negative\'' + ',')
    elif(n < 3 and n >= 1):
        print('\'Positive\'' +',')
    elif(n <= -1 and n > -3):
        print('\'Negative\'' + ',')
    elif(n < 7 and n >= 3):
        print('\'Among top gainers\'' +',')   
    elif(n <= -3 and n > -7):
        print('\'Among top losers\'' + ',')
    elif(n >= 7):
        print('\'Bull run\'' + ',')
    elif(n <= -7):
        print('\'Bear drop\'' + ',')
```

```python
# applying funtion trend to Day_per_change 
df['Trend'] = df[df['Day_Perc_Change'].apply(trend)]
```
